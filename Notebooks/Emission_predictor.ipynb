### Selecting the important features for model

```python
# Feature Engineering
df['Mileage_per_Year'] = df['Mileage'] / (df['Age of Vehicle'] + 1)
df['Vehicle_Age_Group'] = pd.cut(df['Age of Vehicle'],
                                bins=[0, 5, 10, 15, 25, 100],
                                labels=['New', 'Moderate', 'Old', 'Very Old', 'Antique'],
                                include_lowest=True)
```

```python
# Define the set of features based on "Vehicle Characteristics"
vehicle_char_features = [
    'Vehicle Type', 'Fuel Type', 'Engine Size', 'Age of Vehicle',
    'Mileage', 'Speed', 'Acceleration', 'Road Type',
    'Traffic Conditions', 'Temperature', 'Humidity',
    'Wind Speed', 'Air Pressure', 'Mileage_per_Year', 'Vehicle_Age_Group'
]
target = 'Emission Level'
```

```python
# Preprocessing for importance analysis
df_analysis = df[vehicle_char_features + [target]].copy()
```

```python
# Encode Categorical Features
le = LabelEncoder()
cat_cols = ['Vehicle Type', 'Fuel Type', 'Road Type', 'Traffic Conditions', 'Vehicle_Age_Group', target]
for col in cat_cols:
    df_analysis[col] = le.fit_transform(df_analysis[col].astype(str))
```

```python
X = df_analysis.drop(target, axis=1)
y = df_analysis[target]
```

```python
# 1. Random Forest Importance
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
rf_importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
```

```python
# 2. Mutual Information
discrete_features = [X.columns.get_loc(col) for col in X.columns if col in ['Vehicle Type', 'Fuel Type', 'Road Type', 'Traffic Conditions', 'Vehicle_Age_Group']]
mi_scores = mutual_info_classif(X, y, discrete_features=discrete_features, random_state=42)
mi_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)
```

```python
# Visualization
plt.figure(figsize=(12, 12))

plt.subplot(2, 1, 1)
sns.barplot(x=rf_importances.values, y=rf_importances.index, palette='viridis')
plt.title('Feature Importance (Random Forest) - Vehicle Characteristics Only')
plt.xlabel('Importance Score')

plt.subplot(2, 1, 2)
sns.barplot(x=mi_series.values, y=mi_series.index, palette='magma')
plt.title('Mutual Information Scores - Vehicle Characteristics Only')
plt.xlabel('MI Score')

plt.tight_layout()
plt.savefig('vehicle_char_importance.png')
```

1. The Dominance of Emission Features
When all features are included, the model immediately identifies that CO2 Emissions is the primary signal for determining the Emission Level.

CO2 Emissions (Importance ~89%): This feature essentially "defines" the target. In most datasets, "Emission Level" is a categorical binning of CO2 output.

Other Pollutants: While SO2, NOx, PM2.5, and VOC are important, they act as secondary indicators.

Vehicle Characteristics: Features like Engine Size or Fuel Type drop to near-zero relative importance because the model no longer needs to "guess" based on the vehicleit has been given the actual answer (the gas output).
